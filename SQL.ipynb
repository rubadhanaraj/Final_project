```python
import pandas as pd
```

```python
df=pd.read_csv("sales_data.csv")
```

```python
df.head()
```

```python
columns=df.columns
columns
```

```python
sql_df =df.rename({'Row ID': 'Row_ID', 'Order ID': 'Order_ID','Order Date':'Order_Date','Ship Date':'Ship_Date',
                   'Ship Mode':'Ship_Mode','Customer ID':'Customer_ID','Customer Name':'Customer_Name','State':'States',
                   'Postal Code':'Postal_Code','Product ID':'Product_ID','Sub-Category':'Sub_Category',
                   'Product Name':'Product_Name','Shipping Cost':'Shipping_Cost','Order Priority':'Order_Priority',
                   'Per unit sale price':'Per_unit_sale_price','Discount($)':'Discount_dollars',
                   'Per unit discount ($)':'Per_unit_discount_dollars','cost price':'cost_price',
                   'per unit cost price':'per_unit_cost_price',
                   'profit %':'profit_percentage','year':'years','Month':'Months'}, axis='columns')
```

```python
sql_df.head()
```

```python
sql_df.columns
```

```python

#order table

Order=sql_df[["Order_ID","Row_ID","Order_Date","Product_Name","cost_price",'Quantity',"Sales","Discount_dollars"]]
Order.head()
Order.to_csv("Order.csv")
```

```python
#Product table
Product=sql_df[["Order_ID","Product_ID","Category","Sub_Category","Segment"]]
Product.head()
Product.to_csv("Product.csv")
```

```python
#Shipping_info

Shipping_info=sql_df[["Order_ID","City","States","Country","Shipping_Cost",'Order_Priority']]
Shipping_info.head()
Shipping_info.to_csv("Shipping_info.csv")
```

```python
#Customer table

Customer=sql_df[["Product_ID","Customer_Name","Segment"]]
Customer.head()
Customer.to_csv("Customer.csv")
```
