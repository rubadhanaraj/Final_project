```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#load csv and make DataFrame 
sales_df = pd.read_csv('superstore_dataset2011-2015.csv',encoding = "ISO-8859-1")

sales_df.head()
```

```python
# Data type of DataFrame
sales_df.info()
```

```python
#change the "Order Date" and "Ship Date" columns to datetime
sales_df[["Order Date", "Ship Date"]] = sales_df[["Order Date", "Ship Date"]].apply(pd.to_datetime)
sales_df.head()
```

```python
import numpy as np
```

```python
sales_df["Year and Month"]=np.nan
```

```python
sales_df["year and Month"]=sales_df["Ship Date"].apply(lambda x: x.strftime("%Y-%m"))
sales_df.head()
```

```python
# Delete coloumn 
del sales_df["Year and Month"]
```

```python
sales_df.to_csv("clean_sales_csv.csv")
```

```python
sales_df.head()
```

```python
#DataFrame of the sales by category
category_df=sales_df.groupby(["Category","Sub-Category"])["Sales","Quantity"].sum()
category_df
```

```python
# Profit by countries
country_df=sales_df.groupby(["Country"])["Sales","Quantity","Discount","Shipping Cost","Profit"].sum().reset_index()
country_df
```

```python
#DataFrame of the sales by customer segment
customer_segment_df=sales_df.groupby(["Segment"])["Sales","Quantity"].sum()
customer_segment_df
```
