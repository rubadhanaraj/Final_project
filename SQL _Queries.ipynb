```python
# Import dependencies
import psycopg2
import pandas as pd
import numpy as np
import sqlalchemy as sal
from sqlalchemy import create_engine
from config import db_password
db_string = f"postgresql://postgres:{db_password}@127.0.0.1:5432/Superstore_Sales_Data"
engine = create_engine(db_string)
cleaned_superstore_df = pd.read_sql_query ('''
                        SELECT
                        *
                        FROM cleaned_superstore_data
                        ''', engine)
cleaned_superstore_df.head()
```

```python
# customer table
customer_df=cleaned_superstore_df[["Customer_ID","Customer_Name","Segment"]]
# customer_df.head()
customer_df.to_csv("Resources/Customer.csv")
# Convert customer dataframe to SQL database
customer_df.to_sql(name='customer', con=engine,index = False)
```

```python
#sales table
sales_df = cleaned_superstore_df[["Row_ID","Sales","Quantity","Profit","Cost_Price","Months","Years"]]
# sales_df.head()
sales_df.to_csv("Resources/sales.csv")
# Convert customer dataframe to SQL database
sales_df.to_sql(name='sales', con=engine,index = False)
```

```python
#Shipping_info
shipping_info_df = cleaned_superstore_df[["Order_ID","Row_ID","City","States","Country","Shipping_Cost",'Order_Priority',"Ship_Mode","Market","Region"]]
#shipping_info_df.head()
shipping_info_df.to_csv("Resources/Shipping_info.csv")
# Convert shipping dataframe to SQL database
shipping_info_df.to_sql(name='shipping_info', con=engine,index = False)
```

```python
#Product table
product_df = cleaned_superstore_df[["Order_ID","Product_ID","Category","Sub_Category","Segment"]]
#product_df.head()
product_df.to_csv("Resources/Product.csv")
# Convert product dataframe to SQL database
product_df.to_sql(name = 'product', con=engine,index = False)
```
